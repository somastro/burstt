```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## No LNA connection, RRI ANTENNA to Spectrum Analyzer 1-m cable

```python
cols = ['freq_Hz', 'Pow_dBm', 'blank1', 'blank2', 'blank3']
df = pd.read_csv('BURSTT_RRI_GBD1.csv',comment='!', names=cols)
df = df[1:-1]
plt.plot(df['freq_Hz'].astype(float)/1e6, df['Pow_dBm'].astype(float))
plt.xlabel('Frequency (MHz)')
plt.ylabel('Power (dBm)')
plt.title('No LNA connection, RRI ANTENNA-SpAna 1-m cable')
```
